```python
def match_seqs(a,b):
    maxScore = (0, 0, 0)
    for astart, bstart in itertools.product(range(len(a)), range(len(b))):
        astr = a[astart:]
        bstr = b[bstart:]
        score = 0
        strikes = 0
        for matchlen in range(min(len(astr), len(bstr))):
            char1 = astr[matchlen]
            char2 = bstr[matchlen]
            if char1==char2 or 'X' in char1+char2:
                score += 1
                if 'X' in char1+char2:
                    strikes += 1
            else:
                strikes += 1
            if strikes>3:
                break
        if score > maxScore[2]:
            maxScore = (astart, bstart, score)
        if score>40:
            break
    return maxScore
```
